```python
#remover os nan por meio de interpolação polinomial de grau 2

def interpolacao(coluna):
    indices_nan = np.isnan(coluna) #encontrar os indices dos valores nan em booleano
    indices_nao_nan = np.logical_not(indices_nan) #encontrar os indices dos valores NAO nan
    coeficientes = np.polyfit(np.arange(len(coluna))[indices_nao_nan], coluna[indices_nao_nan], deg=2) #funcao que realiza a interpolação, calcula os coeficientes com valores nao nan
    valores_novos = np.polyval(coeficientes, np.arange(len(coluna))) #calcula os valores interpolados para todos os indices
    coluna[indices_nan] = valores_novos[indices_nan] #substitui os valores na coluna original usando os indices booleanos de base
    return coluna
```

```python
train_data['casos'] = interpolacao(train_data['casos'])
train_data['umidmed'] = interpolacao(train_data['umidmed'])
train_data['tempmed'] = interpolacao(train_data['tempmed'])
train_data['tempmax'] = interpolacao(train_data['tempmax'])
train_data['receptivo'] = interpolacao(train_data['receptivo'])
train_data['Rt'] = interpolacao(train_data['Rt'])
train_data['transmissao'] = interpolacao(train_data['transmissao'])
```

```python
import numpy as np
from scipy.interpolate import interp1d
from sklearn.linear_model import LinearRegression

# Defina o tamanho desejado após a interpolação
desired_length = int(np.mean([len(seq) for seq in Xtreino + Xteste]))

def preprocess_sequence(sequence, desired_length):
    if len(sequence) < 2:
        # Preencher com valores fictícios (por exemplo, 0)
        sequence = np.pad(sequence, (0, desired_length - len(sequence)), constant_values=0)
    return sequence


# Função para interpolar sequências
def interpolate_sequence(sequence, desired_length):
    if len(sequence) < 2:
        # Lidar com casos em que a sequência é muito curta
        sequence = preprocess_sequence(sequence, desired_length)
        
    x_original = np.linspace(0, 1, len(sequence))
    x_desired = np.linspace(0, 1, desired_length)
    
    # Interpolação linear (pode mudar para 'quadratic' ou 'cubic')
    f = interp1d(x_original, sequence, kind='linear', fill_value='extrapolate')
    interpolated_sequence = f(x_desired)
    
    return interpolated_sequence

# Aplicar interpolação nas sequências de treino e teste
Xtreino_interpolated = np.array([interpolate_sequence(seq, desired_length) for seq in Xtreino])
Xteste_interpolated = np.array([interpolate_sequence(seq, desired_length) for seq in Xteste])

# Criar e treinar o modelo
modelo = LinearRegression()
modelo.fit(Xtreino_interpolated, Ytreino)

# Fazer previsões
Ptreino = modelo.predict(Xtreino_interpolated)
Pteste = modelo.predict(Xteste_interpolated)
train_model = train_data.copy()
# Se index_treino e index_teste não forem os índices do DataFrame
# Certifique-se de que os índices são compatíveis com os tamanhos dos dados
# train_model.loc[train_model.index[index_treino], 'regressao_linear'] = Ptreino
# train_model.loc[train_model.index[index_teste], 'regressao_linear'] = Pteste

print("Tamanho de Ptreino:", len(Ptreino))
print("Tamanho de index_treino:", len(index_treino))
print("Tamanho de Pteste:", len(Pteste))
print("Tamanho de index_teste:", len(index_teste))

# Exemplo de verificação de comprimento dos índices
print("Comprimento do índice de treino:", len(index_treino))
print("Comprimento do índice de teste:", len(index_teste))

# Certifique-se de que os índices estejam corretos
print("Índices de treino:", index_treino)
print("Índices de teste:", index_teste)

# Exemplo para garantir que o índice booleano tenha o comprimento certo
print("Comprimento do DataFrame:", len(train_model))
print("Comprimento do índice booleano de treino:", len(index_treino))
print("Comprimento do índice booleano de teste:", len(index_teste))

# Verifique quantos dados você está realmente passando para o modelo
print("Tamanho de Xtreino_interpolated:", Xtreino_interpolated.shape)
print("Tamanho de Xteste_interpolated:", Xteste_interpolated.shape)

print("Índices de treino:", index_treino[:10])  # Verifique os primeiros 10 índices
print("Índices de teste:", index_teste[:10])   # Verifique os primeiros 10 índices



cols = ['total_cases', 'regressao_linear']
train_model[cols].plot(figsize = (15,10))
plt.title('Regressao Linear, T=23', fontsize = 25)
```
